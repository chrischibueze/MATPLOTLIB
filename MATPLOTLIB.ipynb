```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_ylabel('Y label')
axes.set_xlabel('X label')
axes.set_title('Set Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.5,0.3])
axes1.plot(x,y)
axes1.set_title('Bigger plot')
axes2.plot(y,x)
axes2.set_title('smaller plot')
```

```python
#legends:  clarify what plot is what plot ...without labels , legends cant work 
```

```python
fig = plt.figure(figsize=(8,4),dpi=100)
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3, label='X cube')
ax.legend(loc=10)  #loc = location of the legend..0 is the best location
```
